```python
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['figure.figsize'] = 8,6
import seaborn as sb
sb.set()
```

```python
amzn = pdr.get_data_stooq('AMZN')
amzn.head()
```

```python
df = amzn['Close']
df_return = round(np.log(df).diff()*100,2)
df_return.head()
```

```python
df_return.plot()
```

```python
df_return[-252:].plot()
```

```python
df_return.dropna(inplace = True)
df_return.describe()
```

```python
from scipy import stats 
stats.describe(df_return)
n,minmax,mean,var,skew,kurt = stats.describe(df_return)
mini,maxi = minmax
std = var**0.5
```

```python
from scipy.stats import norm
```

```python
plt.hist(df_return,bins = 15);
```

```python
x = norm.rvs(mean,std,n) #RVS mean random variable normal distributioj
plt.hist(x,bins=15);
```
